```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Selenium with headless Chrome
options = Options()
options.add_argument("--headless")  # Run in the background
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--log-level=3")  # Suppress logs

# Initialize WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Base URL for pagination
base_url = "https://www.roots.gov.sg/filter/storiesresearch?page={}"

# Function to extract article links
def get_article_links(page_number):
    url = base_url.format(page_number)
    driver.get(url)
    time.sleep(5)  # Allow JavaScript to load

    article_links = []
    
    # Find all story links inside list-item divs
    story_elements = driver.find_elements(By.CSS_SELECTOR, "div.list-item a")

    for story in story_elements:
        link = story.get_attribute("href")
        if link and "/stories-landing/stories/" in link:
            article_links.append(link)

    return article_links

# Scrape all article links from 36 pages
all_article_links = []
for page in range(1, 37):
    print(f"Scraping page {page}...")
    links = get_article_links(page)
    all_article_links.extend(links)

# Close Selenium driver
driver.quit()

# Save links to a file
with open("article_links.txt", "w", encoding="utf-8") as f:
    for link in all_article_links:
        f.write(link + "\n")

print(f"Total articles found: {len(all_article_links)}")
print("Article links saved to article_links.txt")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# Set up Selenium with headless Chrome
options = Options()
options.add_argument("--headless")  
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--log-level=3")  

# Initialize WebDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Read article links from file
with open("article_links.txt", "r", encoding="utf-8") as f:
    all_article_links = [line.strip() for line in f.readlines()]

# Function to extract article text
def get_article_text(article_url):
    driver.get(article_url)
    time.sleep(5)  # Allow JavaScript to load

    try:
        content = driver.find_element(By.CSS_SELECTOR, "div.text-content")  # Change selector if needed
        return content.text.strip()
    except:
        print(f"Failed to extract text from {article_url}")
        return None

# Scrape text from each article
articles_data = []
for idx, link in enumerate(all_article_links, start=1):
    print(f"Scraping article {idx}/{len(all_article_links)}: {link}")
    text = get_article_text(link)
    if text:
        articles_data.append({"url": link, "text": text})

# Close Selenium driver
driver.quit()

# Save results to a CSV file
csv_filename = "roots_sg_articles.csv"
with open(csv_filename, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["url", "text"])
    writer.writeheader()
    writer.writerows(articles_data)

print(f"Scraping completed. Data saved in {csv_filename}")
```

## Data Cleaning of CSV files

```python
# load the full csv and then clean it. store in a pkl file?
import pandas as pd
df=pd.read_csv("roots_sg_articles_full.csv")
df.head()
```

```python
df.info()
```

```python
# clean to remove funny characters
import re
# Clean text: Remove funny characters (e.g., â€™ replacing them with correct apostrophe)
df["text"] = df["text"].str.replace(r"â€™", "'", regex=True)

# Add a new column with "Roots Website" for all rows
df["source"] = "Roots Website"



# Function to extract the title correctly from URL
def extract_title(url):
    if pd.isna(url):  # Check for NaN (missing values)
        return None
    url = str(url)  # Ensure it's a string
    match = re.search(r"/stories/([^/]+)/", url)  # Extract the portion between /stories/ and the next /
    if match:
        title = match.group(1).replace("-", " ").title()  # Convert hyphens to spaces and capitalize
        return title
    return None

# Create new column "title" by extracting it from the URL
df["title"] = df["url"].apply(extract_title)

df.head()
```

```python

# Save the cleaned data
cleaned_csv_filename = "roots_sg_articles_cleaned.csv"
df.to_csv(cleaned_csv_filename, index=False)

print(f"✅ Cleaned data saved to {cleaned_csv_filename}")
```
